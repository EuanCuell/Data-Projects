```python
def dataframe_info(df):
    temp = pd.DataFrame(index=df.columns)
    temp["Datatype"] = df.dtypes
    temp["Not null values"] = df.count()
    temp["Null values"] = df.isnull().sum()
    temp["Percentage of Null values"] = (df.isnull().mean())*100
    temp["Unique count"] = df.nunique()
    return temp
dataframe_info(data)
```

# Drop unnecessary column

```python
data = data.drop(['index'], axis=1)
data.head()
```

# Min, Max, Avg Salary Estimate

```python
print(data['Salary Estimate'].unique())
```

```python
# Remove the '$', 'K', and the unecessary string
temp = data['Salary Estimate'].str.split('(')
salary_estimate = []
for x in temp:
    x = x[0]
    salary_estimate.append(x)
data['Salary Estimate'] = salary_estimate
data['Salary Estimate'] = data['Salary Estimate'].str.replace('K', '')
data['Salary Estimate'] = data['Salary Estimate'].str.replace('$', '')
print(data['Salary Estimate'])
```

```python
# Extract the minimum and maximum salary, and calculat the average salary
temp = data['Salary Estimate'].str.split('-')

min_salary = []
for x in temp:
    x = x[0]
    min_salary.append(x)
data['Min Salary'] = min_salary
data['Min Salary'] = data['Min Salary'].astype(int)

max_salary = []
for x in temp:
    x = x[1]
    max_salary.append(x)
data['Max Salary'] = max_salary
data['Max Salary'] = data['Max Salary'].astype(int)

data['Avg Salary'] = (data['Min Salary'] + data['Max Salary']) // 2

data[['Salary Estimate', 'Min Salary', 'Max Salary', 'Avg Salary']]
```

# Company Name

```python
# Company names have their ratings beside them
print(data['Company Name'])
```

```python
# Separate the company name and rating 
temp_1 = data['Company Name'].str.split('\n')
print(temp_1)
company_name = []
for x in temp_1:
    x = x[0]
    company_name.append(x)
data['Company Name'] = company_name
data['Company Name']
```

# Location and State

```python
data['Location']
```

```python
# Isolate the state from the location 
temp = data['Location'].str.split(', ')

state = []
for x in temp:
    x = x[-1] # Have to use -1 instead of 1, because not all values in 'Location' are in the 'city, state' format, some are just 'Remote', so when the split is applied to that row, nothing happens. So in temp, some cells look like ['New York', 'NY'], and some look like ['Remote'], so when iterating over temp, if we selected x[1], there would be an error, cause whilst for some rows x[1] would select the state appropriately 'NY', for the rows that are just ['Remote'], x[1] isn't an index. But if we use x[-1], it would still select the state appropriately, AND would also select ['Remote'] appropriately cause x[-1] is the same as x[0] 
    state.append(x)
    
data['State'] = state
data[['Location', 'State']]
```

```python
#'State' column contains some irregular values,
print(data['State'].value_counts())

# Convert those irregular values to the appropriate format
state = []
for x in data['State']:
    if 'United States' in x:
        state.append('N/A')
    elif 'Remote' in x:
        state.append('N/A')
    elif 'New Jersey' in x:
        state.append('NJ')
    elif 'Utah' in x:
        state.append('UT')
    elif 'Texas' in x:
        state.append('TX')
    elif 'California' in x:
        state.append('CA')
    else:
        state.append(x)
data['State'] = state
print(data['State'].value_counts())
```

```python
# Create 'Same State' column if the 'Location' is the same as the 'Headquarters'
data['Same State'] = np.where(data['Location'] == data['Headquarters'], 1, 0)
data[['Location', 'Headquarters', 'Same State']]
```

# Year Founded -> Company Age

```python
print(data['Founded'])
```

```python
# Calculate company age from year founded
company_age = []
for x in data['Founded']:
    if x < 0:
        company_age.append(x)
    else:
        company_age.append(2023-x)
data['Company Age'] = company_age
data[['Founded', 'Company Age']]
        
```

# Rating

```python
print(data['Rating'].unique())
```

```python
# Convert inappropriate ratings to 0
data['Rating'] = np.where(data['Rating'] == -1, 0, data['Rating'])
print(data['Rating'].unique())
```

# Job Title

```python
print(data['Job Title'].value_counts())
```

```python
data['Job Title'] = data['Job Title'].str.lower()
print(data['Job Title'].unique())
```

```python
# Simplify 'Job Title'
job_simp = []

for x in data['Job Title']:
    if 'data scientist' in x:
        job_simp.append('data scientist')
    elif 'data engineer' in x:
        job_simp.append('data engineer')
    elif 'machine learning' in x:
        job_simp.append('mle')
    elif 'analyst' in x:
        job_simp.append('analyst')
    elif 'manager' in x:
        job_simp.append('manager')
    elif 'director' in x:
        job_simp.append('director')
    else:
        job_simp.append('na')
data['Job Simplified'] = job_simp
data[['Job Title', 'Job Simplified']]
```

# Job Seniority

```python
# Assign job seniority status based on keywords
seniority = []

for x in data['Job Title']:
    if ('sr' in x) or ('senior' in x) or ('lead' in x) or ('sr.' in x) or ('manager' in x):
        seniority.append('Senior')
    elif ('jr' in x) or ('jr.' in x):
        seniority.append('Junior')
    else:
        seniority.append('N/A')
data['Seniority'] = seniority
data[['Job Title', 'Job Simplified', 'Seniority']]
```

```python
data['Job Description'] = data['Job Description'].str.lower()

# Extract keywords from the 'Job Description'
python = []
for x in data['Job Description']:
    if 'python' in x:
        python.append(1)
    else:
        python.append(0)
data['python'] = python

excel = []
for x in data['Job Description']:
    if 'excel' in x:
        excel.append(1)
    else:
        excel.append(0)
data['excel'] = excel

hadoop = []
for x in data['Job Description']:
    if 'hadoop' in x:
        hadoop.append(1)
    else:
        hadoop.append(0)
data['hadoop'] = hadoop

spark = []
for x in data['Job Description']:
    if 'spark' in x:
        spark.append(1)
    else:
        spark.append(0)
data['spark'] = spark

aws = []
for x in data['Job Description']:
    if 'aws' in x:
        aws.append(1)
    else:
        aws.append(0)
data['aws'] = aws

tableau = []
for x in data['Job Description']:
    if 'tableau' in x:
        tableau.append(1)
    else:
        tableau.append(0)
data['tableau'] = tableau

big_data = []
for x in data['Job Description']:
    if 'big_data' in x:
        big_data.append(1)
    else:
        big_data.append(0)
data['big_data'] = big_data

data[['python', 'excel', 'hadoop', 'spark', 'aws', 'tableau', 'big_data']]
```
